```python
import sqlite3
conn = sqlite3.connect('task1_7.db')
```

```python
import random
```

```python
import pandas as pd
```

```python
data = [[random.randrange(0, 1000) for j in range(3)] for i in range(1000)]
```

```python
cur = conn.cursor()
```

```python
cur.execute("CREATE TABLE table1(int_1, int_2, int_3)")
```

```python
cur.executemany("insert into table1 values(?, ?, ?)", data)
conn.commit()
```

```python
query = 'select * from table1'
df = pd.read_sql(sql=query, con=conn)
df
```

```python
import seaborn as sns
```

```python
# sns.displot(data=tips, x="total_bill", col="time", kde=True)
sns.displot(data=df, x="int_1", kde=True)
sns.displot(data=df, x="int_2", kde=True)
sns.displot(data=df, x="int_3", kde=True)
```
